```python
import cv2
from deepface import DeepFace

# Yüz algılayıcıyı yükle
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Kameradan görüntü al
cap = cv2.VideoCapture(0)

while True:
    # Kameradan kareyi oku
    ret, frame = cap.read()
    
    # Görüntüyü gri tonlamaya çevir
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Yüzleri algıla
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    dominant_emotion = "No Emotion Detected"  # Varsayılan ifade
    for (x, y, w, h) in faces:
        # Yüzün çevresine dikdörtgen çiz
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        
        # Yüz bölgesini kırp
        face = frame[y:y+h, x:x+w]
        
        # Eğer yüz yeterince büyükse analiz yap
        if w > 50 and h > 50:  # Yüzün boyutunu kontrol et
            try:
                emotion = DeepFace.analyze(face, actions=['emotion'])
                dominant_emotion = emotion['dominant_emotion']  # Tespit edilen yüz ifadesi
                
                # Yüz ifadelerine göre mesaj belirle
                if dominant_emotion == "happy":
                    dominant_emotion = "Mutlu"
                elif dominant_emotion == "sad":
                    dominant_emotion = "Üzgün"
                elif dominant_emotion == "surprise":
                    dominant_emotion = "Şaşkın"
                elif dominant_emotion == "angry":
                    dominant_emotion = "Kızgın"
                else:
                    dominant_emotion = dominant_emotion.capitalize()  # Diğer ifadeler
            except Exception as e:
                print(f"Hata: {e}")  # Hata varsa yazdır
                dominant_emotion = "Emotion not detected"
    
    # Yüz ifadesini sol üst köşeye yazdır
    cv2.putText(frame, dominant_emotion, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

    # Hareket algılama
    if 'previous_frame' in locals():
        frame_diff = cv2.absdiff(previous_frame, gray)
        _, thresh = cv2.threshold(frame_diff, 25, 255, cv2.THRESH_BINARY)
        contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        
        for contour in contours:
            if cv2.contourArea(contour) > 500:  # Küçük hareketleri ihmal et
                (x, y, w, h) = cv2.boundingRect(contour)
                cv2.rectangle(frame, (x, y, w, h), (0, 0, 255), 2)
    
    # Şu anki kareyi arka plan referansı için sakla
    previous_frame = gray.copy()

    # Sonucu göster
    cv2.imshow('Hareket ve Yüz İfadesi Algılama', frame)
    
    # 'q' tuşuna basarak çık
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Kaynakları serbest bırak
cap.release()
cv2.destroyAllWindows()
```
